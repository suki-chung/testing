```python
import plotly.express as px
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(42)

data = {
    'Category': np.random.choice(['A', 'B', 'C'], size=100),
    'Value': np.random.rand(100) * 100,  # Random values between 0 and 100
    'X': np.random.rand(100) * 100,  # Random values for X-axis
    'Y': np.random.rand(100) * 100,  # Random values for Y-axis
    'Location': [f"Location {i+1}" for i in range(100)],
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create a scatter plot with Plotly Express
fig = px.scatter(df, x='X', y='Y', color='Category', size='Value', hover_name='Location',
                 title="Random Interactive Scatter Plot")

# Show the plot
fig.show()
```
